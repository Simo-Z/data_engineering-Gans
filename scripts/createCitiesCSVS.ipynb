```python
import pandas as pd
```

```python
# https://simplemaps.com/data/world-cities
cities = pd.read_csv("../data/worldcities.csv")
```

```python
german_cities = cities.loc[cities["iso2"] == "DE"].reset_index(drop=True)
german_cities
```

```python
Berlin = german_cities.loc[german_cities["city"] == "Berlin"]
Berlin
```

```python
Berlin.to_csv("../csv/Berlin_city_data.csv", index=False)
cities.to_csv("../csv/world_city_data.csv", index=False)
german_cities.to_csv("../csv/germany_city_data.csv", index=False)
```
