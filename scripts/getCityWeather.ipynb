```python
# import libraries
import requests
import pandas as pd
import time
import chime
import os
from dotenv import load_dotenv, find_dotenv


chime.theme("big-sur")
```

```python
# Keys
load_dotenv(find_dotenv(filename='.env'))
API_KEY = os.environ["OWM_API_KEY"]
```

```python
# get cities
german_citys = pd.read_csv("../data/csv/germany_city_data.csv")
big_german_citys = german_citys.loc[german_citys["population"] > 50000]
big_german_citys
```

```python
def get_city_weather_dict(city_weather_json, city_row):
    city_weather_dict = {
            "city_name": city_row["city_ascii"],
            "country_code": city_weather_json["sys"]["country"],
            "w_main": city_weather_json["weather"][0]["main"],
            "w_description": city_weather_json["weather"][0]["description"],
            "w_icon": city_weather_json["weather"][0]["icon"],
            "w_temp": city_weather_json["main"]["temp"],
            "w_feels_like": city_weather_json["main"]["feels_like"],
            "w_temp_min": city_weather_json["main"]["temp_min"],
            "w_temp_max": city_weather_json["main"]["temp_max"],
            "w_pressure": city_weather_json["main"]["pressure"],
            "humidity": city_weather_json["main"]["humidity"],
            "wind_speed": city_weather_json["wind"]["speed"],
            "wind_deg": city_weather_json["wind"]["deg"],
            "clouds": city_weather_json["clouds"]["all"],
            "dt": city_weather_json["dt"],
            "sunrise": city_weather_json["sys"]["sunrise"],
            "sunset": city_weather_json["sys"]["sunset"],
            "timezone": city_weather_json["timezone"],
            "w_name": city_weather_json["name"]
        }
    return city_weather_dict
```

```python
city_weather_fc = pd.DataFrame()
city_list = big_german_citys[["city_ascii", "lat", "lng"]]
```

```python
chime.info()
for index, row in city_list.iterrows():
    url = f"https://api.openweathermap.org/data/2.5/weather/?lat={row['lat']}&lon={row['lng']}&appid={API_KEY}"
    try: 
        response = requests.get(url).json()
        
        city_weather_fc = city_weather_fc.append({
            "city_name": row["city_ascii"],
            "country_code": response["sys"]["country"],
            "w_main": response["weather"][0]["main"],
            "w_description": response["weather"][0]["description"],
            "w_icon": response["weather"][0]["icon"],
            "w_temp": response["main"]["temp"],
            "w_feels_like": response["main"]["feels_like"],
            "w_temp_min": response["main"]["temp_min"],
            "w_temp_max": response["main"]["temp_max"],
            "w_pressure": response["main"]["pressure"],
            "humidity": response["main"]["humidity"],
            "wind_speed": response["wind"]["speed"],
            "wind_deg": response["wind"]["deg"],
            "clouds": response["clouds"]["all"],
            "dt": response["dt"],
            "sunrise": response["sys"]["sunrise"],
            "sunset": response["sys"]["sunset"],
            "timezone": response["timezone"],
            "w_name": response["name"]
        }, ignore_index=True)
    except Exception as e:
        print(f"Error {row['city_ascii']}")
        raise e
chime.success()
city_weather_fc
```

```python
city_weather_fc.to_csv("../final_tables/german_city_weather_fc.csv", index=False)
```
